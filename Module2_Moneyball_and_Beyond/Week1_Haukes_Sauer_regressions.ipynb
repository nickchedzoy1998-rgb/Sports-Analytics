```python
cols_to_aggregate = [
    'home_ab', 'home_h', 'home_2b', 'home_3b', 'home_hr',
    'home_bb', 'home_hbp', 'home_sf',
    'visitor_ab', 'visitor_h', 'visitor_2b', 'visitor_3b',
    'visitor_hr', 'visitor_bb', 'visitor_hbp', 'visitor_sf',
    'Hwin', 'Awin'
]

for col in cols_to_aggregate:
    if col in list(df.columns):
        continue
    else: print(f'{col} not found in df')
```

```python
home_df = df.groupby(['year', 'home'])[cols_to_aggregate].sum().reset_index()
home_df = home_df.rename(columns={'home': 'Team'})

visitor_df = df.groupby(['year', 'visitor'])[cols_to_aggregate].sum().reset_index()
visitor_df = visitor_df.rename(columns={'visitor': 'Team'})
```

```python
# Season Table

season_df = pd.merge(home_df, visitor_df, on=['year', 'Team'])
season_df
```

```python
# Sanity Check

check_col_string = 'ab'
check_cols = []

for col in season_df.columns:
    if check_col_string in col:
        check_cols.append(col)

season_df[check_cols]
```

```python
for_metrics = [('home', 'x'), ('visitor', 'y')]
against_metrics = [('home', 'y'), ('visitor', 'x')]
rename_map = {}

for col in season_df.columns:
    if 'home' not in str(col) and 'visitor' not in str(col):
        continue
    
    if for_metrics[0][0] in col and for_metrics[0][1] in col:
        new_name = str(col).replace('x', 'for')
    if for_metrics[1][0] in col and for_metrics[1][1] in col:
        new_name = str(col).replace('y', 'for')

    if against_metrics[0][0] in col and against_metrics[0][1] in col:
        new_name = str(col).replace('y', 'against')
    if against_metrics[1][0] in col and against_metrics[1][1] in col:
        new_name = str(col).replace('x', 'against')

    rename_map[col] = new_name

keys_to_delete = []

for key in rename_map.keys():
    if 'home' not in str(key) and 'visitor' not in str(key):
        keys_to_delete.append(str(key))

for key in keys_to_delete:
    del rename_map[key]

rename_map

season_df = season_df.rename(columns=rename_map)

season_df
```

```python
wins = ['Hwin_x', 'Awin_y']
loss = ['Hwin_y', 'Awin_x']

season_df = season_df.rename(
    columns = {
        'Hwin_x': 'Hwin', 
        'Awin_y': 'Awin', 
        'Hwin_y': 'Aloss',
        'Awin_x': 'Hloss'})

season_df['Total_Wins'] = season_df['Hwin'] + season_df['Awin']
season_df['Total_Losses'] = season_df['Hloss'] + season_df['Aloss']

season_df
```

```python
season_df['Total_Games'] = season_df['Total_Wins'] + season_df['Total_Losses']
```

```python
season_df['W_perc'] = season_df['Total_Wins'] / season_df['Total_Games']
```

```python
# OBP: On Base Percentage

OBP_FOR_COLS = [
    'home_h_x', 
    'visitor_h_y', 
    'home_bb_x', 
    'visitor_bb_y', 
    'home_hbp_x', 
    'visitor_hbp_y',
    'home_ab_x', 
    'visitor_ab_y', 
    'home_bb_x', 
    'visitor_bb_y', 
    'home_hbp_x', 
    'visitor_hbp_y', 
    'home_sf_x', 
    'visitor_sf_y'
]

# OBP FOR FORMULA:
"""
('home_h_x' + 'visitor_h_y' + 'home_bb_x' + 
'visitor_bb_y' + 'home_hbp_x' + 'visitor_hbp_y')
/
('home_ab_x' + 'visitor_ab_y' + 'home_bb_x' + 'visitor_bb_y' + 
'home_hbp_x' + 'visitor_hbp_y' + 'home_sf_x' + 'visitor_sf_y')
"""

for i, col in enumerate(OBP_FOR_COLS):
    new = rename_map[col]
    OBP_FOR_COLS[i] = new
```

```python
# First 6 cols in first part of bracket formula, Remaining 8 in second part - Check:
print(f'{OBP_FOR_COLS[0:6]}\n{OBP_FOR_COLS[6:]}')
```

```python
dividend_mask = OBP_FOR_COLS[0:6]
divisor_mask = OBP_FOR_COLS[6:]

season_df['OBP_FOR'] = (season_df[dividend_mask].sum(axis=1)) / (season_df[divisor_mask].sum(axis=1))
season_df['OBP_FOR']
```

```python
OBP_AGN_COLS = [
    'home_h_y', 'visitor_h_x', 'home_bb_y', 
    'visitor_bb_x', 'home_hbp_y', 'visitor_hbp_x',
    'home_ab_y', 'visitor_ab_x', 'home_bb_y', 'visitor_bb_x', 
    'home_hbp_y', 'visitor_hbp_x', 'home_sf_y', 'visitor_sf_x']

# OBP_AGN Formula
"""
('home_h_y' + 'visitor_h_x' + 'home_bb_y' + 
'visitor_bb_x' + 'home_hbp_y' + 'visitor_hbp_x')
/
('home_ab_y' + 'visitor_ab_x' + 'home_bb_y' + 'visitor_bb_x' + 
'home_hbp_y' + 'visitor_hbp_x' + 'home_sf_y' + 'visitor_sf_x')

"""

for i, col in enumerate(OBP_AGN_COLS):
    new = rename_map[col]
    OBP_AGN_COLS[i] = new

dividend_mask = OBP_AGN_COLS[0:6]
divisor_mask = OBP_AGN_COLS[6:]

season_df['OBP_AGN'] = (season_df[dividend_mask].sum(axis=1)) / (season_df[divisor_mask].sum(axis=1))
season_df.iloc[:, -2:]
```

```python
# SLG: Slugging Percentage:
# The average number of bases the team gains per at-bat

# Formula:
"""
((home_h_x + visitor_h_y) - (home_2b_x + visitor_2b_y) - 
(home_3b_x + visitor_3b_y) - (home_hr_x + visitor_hr_y) +

    2 * (home_2b_x + visitor_2b_y) +
    3 * (home_3b_x + visitor_3b_y) +
    4 * (home_hr_x + visitor_hr_y) ) 
/ 
(home_ab_x + visitor_ab_y)

"""

SLG_FOR_COLS = [
    'home_h_x',
    'visitor_h_y',
    'home_2b_x',
    'visitor_2b_y',
    'home_3b_x',
    'visitor_3b_y',
    'home_hr_x',
    'visitor_hr_y',
    'home_2b_x',
    'visitor_2b_y',
    'home_3b_x',
    'visitor_3b_y',
    'home_hr_x',
    'visitor_hr_y',
    'home_ab_x',
    'visitor_ab_y'
    ]

for i, col in enumerate(SLG_FOR_COLS):
    new_name = rename_map[col]
    SLG_FOR_COLS[i] = new_name

pos_dict = {}

# For ease of use print cols with table name:
for i, col in enumerate(SLG_FOR_COLS):
    pos_dict[i] = f"""season_df['{col}']"""


print(f"""
(({pos_dict[0]} + {pos_dict[1]}) - ({pos_dict[2]} + {pos_dict[3]}) - 
({pos_dict[4]} + {pos_dict[5]}) - ({pos_dict[6]} + {pos_dict[7]}) +
    2 * ({pos_dict[8]} + {pos_dict[9]}) +
    3 * ({pos_dict[10]} + {pos_dict[11]}) +
    4 * ({pos_dict[12]} + {pos_dict[13]}) ) 
/ 
({pos_dict[14]} + {pos_dict[15]})
      """)
```

```python
season_df['SLG_FOR'] = (
    (season_df['home_h_for'] + season_df['visitor_h_for']) - (season_df['home_2b_for'] + season_df['visitor_2b_for']) - 
    (season_df['home_3b_for'] + season_df['visitor_3b_for']) - (season_df['home_hr_for'] + season_df['visitor_hr_for']) +
    2 * (season_df['home_2b_for'] + season_df['visitor_2b_for']) +
    3 * (season_df['home_3b_for'] + season_df['visitor_3b_for']) +
    4 * (season_df['home_hr_for'] + season_df['visitor_hr_for'])) /(season_df['home_ab_for'] + season_df['visitor_ab_for'])
```

```python
season_df['SLG_AGN'] = (
    (season_df['home_h_against'] + season_df['visitor_h_against']) - (season_df['home_2b_against'] + season_df['visitor_2b_against']) - 
    (season_df['home_3b_against'] + season_df['visitor_3b_against']) - (season_df['home_hr_against'] + season_df['visitor_hr_against']) +
    2 * (season_df['home_2b_against'] + season_df['visitor_2b_against']) +
    3 * (season_df['home_3b_against'] + season_df['visitor_3b_against']) +
    4 * (season_df['home_hr_against'] + season_df['visitor_hr_against'])) /(season_df['home_ab_against'] + season_df['visitor_ab_against'])
```

```python
"""
Create batting average for and batting average against as the sum the relevant statistics for each team as home team and visiting team.

Create the win percentage variable for each team by generating a count of games played and dividing total wins by total games played.

Using a similar process to part 2), calculate each team’s home and away win percentage separately for each season. 

Merge each team’s home and away win percentage for each season into one data frame. Calculate WPCT diff defined as:
- home win percentage –
- away win percentage for every team in each season
"""
```

```python
# Hits / AtBats

season_df['BA_for'] = (season_df['home_h_for'] + season_df['visitor_h_for']) / (season_df['home_ab_for'] + season_df['visitor_ab_for'])
season_df['BA_against'] = (season_df['home_h_against'] + season_df['visitor_h_against']) / (season_df['home_ab_against'] + season_df['visitor_ab_against'])
```

```python
# Home + Away Win pct

season_df['HWin_pct'] = season_df['Hwin'] / (season_df['Hwin'] + season_df['Hloss'])
season_df['AWin_pct'] = season_df['Awin'] / (season_df['Awin'] + season_df['Aloss'])
```

```python
for col in season_df.columns:
    if str.lower('win') in col or str.lower('loss') in col:
        print(col)
```

```python
season_df['WPCT_diff'] = season_df['HWin_pct'] - season_df['AWin_pct']
```

```python
# Question 1
# Find the team with the maximum single season WPCT diff in the dataframe. What was this team’s away winning percentage for the season?

q1 = season_df.loc[season_df['WPCT_diff'] == season_df['WPCT_diff'].max(), 'AWin_pct']
q1
```

```python
# Question 2
# Create a variable AVG Diff = 
# Batting Average For – Batting Average Against. What was the minimum (most negative) single season team AVG Diff in the dataframe?

season_df['BA_diff'] = season_df['BA_for'] - season_df['BA_against']
season_df['BA_diff'].min()
```

```python
# Assignment Part 3:

"""
Run an OLS (Ordinary Least Squares) regression of win percentage on batting average for and batting average against.

Run an OLS (Ordinary Least Squares) regression of win percentage on batting average for and batting average against,   
constraining the coefficient on each variable to be equal and opposite."""
```

```python
import statsmodels.formula.api as smf

WinBA_lm = smf.ols(formula = 'W_perc ~ BA_for + BA_against', data=season_df).fit()
WinBA_lm.summary()
```

```python
import statsmodels.formula.api as smf

WinBA_lm_constrained = smf.ols(formula = 'W_perc ~ BA_diff', data=season_df).fit()
WinBA_lm_constrained.summary()
```
