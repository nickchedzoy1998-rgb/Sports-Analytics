## Part 2 - Conditional Probability and Autocorrelation

```python
# Indicate where a player had a hit on consecutive shots

Shotlog_1415['cons_shot_hit'] = np.where(
    (Shotlog_1415['current_shot_hit'] == 1) & (Shotlog_1415['lag_shot_hit'] == 1),
    1,
    0
)
```

```python
# Calculate Average cons_hit and Average lag_hit

player_prob = (
    Shotlog_1415
    .groupby('shoot_player')[['cons_shot_hit', 'lag_shot_hit']]
    .mean()
    .reset_index()
)
```

```python
# Conditional Probability

player_prob['conditional_prob'] = player_prob['cons_shot_hit'] / player_prob['lag_shot_hit']
player_prob
```

```python
Player_Stats = pd.merge(player_prob, Player_Stats, on = ['shoot_player'])
Player_Stats
```

```python
Player_Stats.info()
```

```python
# Drop Vals where average lag shot = 0

Player_Stats = Player_Stats[pd.notnull(Player_Stats['conditional_prob'])]
```

```python
Player_Stats = Player_Stats.sort_values(by=['conditional_prob'], ascending=[False])
Player_Stats
```

```python
# find the difference in probabilities for each player

Player_Stats['prob_dif'] = Player_Stats['conditional_prob'] - Player_Stats['average_hit']
```

```python
# add shot count to df

player_shots = Shotlog_1415[['shoot_player', 'shot_count']]
Player_Stats = pd.merge(Player_Stats, player_shots, on=['shoot_player'])
Player_Stats = Player_Stats.sort_values(by=['prob_dif'], ascending = [False])
Player_Stats
```

```python
Player_Stats
```

```python
# t-test to test if the players’ probability of hit is statistically significantly different than their conditional probability

import scipy.stats as sp

sp.stats.ttest_ind(Player_Stats['conditional_prob'], Player_Stats['average_hit'])
```

```python
## Autocorrelation Coefficient
# Calculate the autocorrelation coefficient “current_shot_hit” variable and the “lag_shot_hit” variable. 

Shotlog_1415['current_shot_hit'].corr(Shotlog_1415['lag_shot_hit'])
```

```python
# Calculate coefficient for each player

player_coef = Shotlog_1415.groupby('shoot_player')[['current_shot_hit', 'lag_shot_hit']].corr().unstack().iloc[:,1].reset_index()
player_coef.columns = player_coef.columns.get_level_values(0)
player_coef = player_coef.rename(columns = {'current_shot_hit':'autocor'})
player_coef
```

```python
avg_player_shots = Shotlog_1415[['shoot_player', 'shot_per_game']]
avg_player_shots = avg_player_shots.groupby(['shoot_player'])['shot_per_game'].mean().reset_index(name='avg_shot_game')
avg_player_shots
```

```python
player_coef = pd.merge(player_coef, avg_player_shots, on = 'shoot_player').sort_values(by=['autocor'], ascending=[False])
player_coef
```

## Part 3 - Regression Analyses

```python
# prediction error : dif betwen outcome and average success rate

Shotlog_1415['error'] = Shotlog_1415['current_shot_hit'] - Shotlog_1415['average_hit']
Shotlog_1415['lag_error'] = Shotlog_1415['lag_shot_hit'] - Shotlog_1415['average_hit']
```

```python
Shotlog_1415['error'].describe()
```

```python
Shotlog_1415['lag_error'].describe()
```

```python
# linear regression controlling for variables

import statsmodels.api as sm
import statsmodels.formula.api as smf

reg_1 = smf.ols(
    formula='error ~ lag_error + shot_dist + dribbles + touch_time + C(points) + C(quarter) + home_away + closest_def_dist + C(shoot_player) + C(closest_defender)',
    data=Shotlog_1415
).fit()

print(reg_1.summary())
```

```python
# Weighted Least Squares Regression by no shots/game

reg_2 = smf.wls(
    formula='error ~ lag_error + shot_dist + dribbles + touch_time + C(points) + C(quarter) + home_away + closest_def_dist + C(shoot_player) + C(closest_defender)',
    data=Shotlog_1415,
    weights=1/Shotlog_1415['shot_per_game']
).fit()

print(reg_2.summary())
```

```python
import statsmodels.formula.api as smf

players_to_run = [
    "russell westbrook",
    "james harden",
    "andrew wiggins",
    "stephen curry"
]

reg3_results = {}

for p in players_to_run:
    df_p = Shotlog_1415[Shotlog_1415['shoot_player'] == p].copy()

    # Skip if too little data
    if df_p.shape[0] < 30:
        print(f"{p}: Not enough data ({df_p.shape[0]} shots)")
        continue

    # Drop constant columns (no variation → regression kill switch)
    for col in ['points','quarter','home_away']:
        if df_p[col].nunique() < 2:
            df_p.drop(columns=[col], inplace=True)

    formula = 'error ~ lag_error + shot_dist + dribbles + touch_time + closest_def_dist'

    # Add categorical predictors only if variation exists
    if 'points' in df_p.columns:
        formula += ' + C(points)'
    if 'quarter' in df_p.columns:
        formula += ' + C(quarter)'
    if 'home_away' in df_p.columns:
        formula += ' + home_away'

    print(f"\n\n===== {p} =====")
    model = smf.ols(formula=formula, data=df_p).fit()
    reg3_results[p] = model
    print(model.summary())
```
